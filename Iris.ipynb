Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
```

```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
std_scaler=StandardScaler()
X=std_scaler.fit_transform(X)
```

```python
X_Train, X_Test, Y_Train, Y_Test= train_test_split(X,Y,test_size=0.2, random_state=7)
```

Logistic regression


```python
logreg.fit(X_Train,Y_Train)
ylog_pred=logreg.predict(X_Test)
```

```python
sns.heatmap(confusion_matrix(Y_Test,ylog_pred),annot=True)
plt.xlabel("Predicted label")
plt.ylabel("True label")
```

```python
print(logreg.score(X_Test,Y_Test))
```

```python
print(logreg.predict(std_scaler.transform([[4.9,3.5,1.6,0.5]])))
```

```python
print(logreg.predict(std_scaler.transform([[5.8,2.1,4.1,1.4]])))
```

```python
ts=0.1
sizes=[]
scores=[]
for i in range(1,10):
  X_Train, X_Test, Y_Train, Y_Test= train_test_split(X,Y,test_size=ts)
  logreg.fit(X_Train,Y_Train)
  scores.append(logreg.score(X_Test,Y_Test))
  sizes.append(ts)
  ts=ts+0.1

plt.plot(sizes,scores)
plt.xlabel("Test sizes")
plt.ylabel("Accuracy score")
plt.show()

```

KNN

```python
train_accuracy=[]
test_accuracy=[]
no_neighbors=[]
for i in range (1,13):
  classifier=KNeighborsClassifier(n_neighbors=i).fit(X_Train,Y_Train)
  no_neighbors.append(i)
  test_accuracy.append(classifier.score(X_Test,Y_Test))
  train_accuracy.append(classifier.score(X_Train,Y_Train))
```

```python
plt.figure(figsize=(10,8))
plt.plot(no_neighbors,test_accuracy,color='r', label='Testing accuracy')
plt.plot(no_neighbors,train_accuracy,color='b', label='Training accuracy')
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
plt.legend()
plt.show()
```

```python
classifier=KNeighborsClassifier(n_neighbors=7)
```

```python
classifier.fit(X_Train,Y_Train)
Y_Pred=classifier.predict(X_Test)
```

```python
sns.heatmap(confusion_matrix(Y_Test,Y_Pred),annot=True)
plt.xlabel("Predicted label")
plt.ylabel("True label")
```

```python
print(classifier.score(X_Test,Y_Test))
```
